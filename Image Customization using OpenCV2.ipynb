# Image Customization using OpenCV2:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
plt.figure(figsize=(5,3))
img = plt.imread("per_cat.jpg")
plt.imshow(img)
plt.grid(False)
plt.show()
```

```python
img.shape
```

```python
def catImageshow(imageTitle,image):
    color_convert = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(5,3))
    plt.imshow(color_convert)
    plt.title(imageTitle)
    plt.show()
catImageshow("Original Image",img)
```

### 1. Adding Gray Effect to Image:

```python
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
gray.shape
```

```python
catImageshow("Gray Image",gray)
```

```python
gradientX = cv2.Sobel(gray,ddepth = cv2.CV_32F,dx=1,dy=0,ksize=3)
gradientY = cv2.Sobel(gray,ddepth = cv2.CV_32F,dx=0,dy=1,ksize=3)
```

```python
catImageshow("GradientX",gradientX)
```

```python
catImageshow("GradientY",gradientY)
```

### 2. Masking the Image:

```python
img.shape
```

```python
img.shape[:2]
```

```python
mask = np.zeros(img.shape[:2],dtype="uint8")
```

```python
print(mask)
```

```python
cv2.rectangle(mask,(100, 500), (150,100), 0)
catImageshow("Mask Image",mask)
```

### 3. Bit Mask Image:

```python
cv2.circle(mask, (350,200), 140, 250, -1)
bit_mask = cv2.bitwise_and(img,img,mask = mask)
catImageshow("Mask Image",mask)
catImageshow("Bit_Mask Image",bit_mask)
```

### 5. Resize Image:

```python
customSizeH = 120/img.shape[0]
```

```python
customSizeW = 120/img.shape[1]
```

```python
print(customSizeW)
```

```python
print(customSizeH)
```

```python
imgDimension = (120,int(img.shape[0]*customSizeW))
```

```python
print(imgDimension)
```

```python
img.shape
```

```python
resizeImage = cv2.resize(img,imgDimension,interpolation = cv2.INTER_AREA)
catImageshow("Resized Image",resizeImage)
```

```python
resizeImage.shape
```

### 6. Rotating the Image:

```python
(imageH,imageW) = img.shape[:2]
```

```python
print(imageH)
```

```python
print(imageW)
```

```python
(centerX,centerY) = (imageH//2,imageW//2)
```

```python
print(centerX)
```

```python
print(centerY)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),45,1.0)
rotateNow = cv2.warpAffine(img,imageRotate, (imageH,imageW))
catImageshow("Rotate 45 Degree",rotateNow)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),90,1.0)
rotateNow = cv2.warpAffine(img,imageRotate, (imageH,imageW))
catImageshow("Rotate 90 Degree",rotateNow)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),120,1.0)
rotateNow = cv2.warpAffine(img,imageRotate, (imageH,imageW))
catImageshow("Rotate 120 Degree",rotateNow)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),180,1.0)
rotateNow = cv2.warpAffine(img,imageRotate, (imageH,imageW))
catImageshow("Rotate 180 Degree",rotateNow)
```
